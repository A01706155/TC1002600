```python
#Importar datos
import pandas as pd
import statistics as st
from matplotlib import pyplot as plt
data = pd.read_csv("artists_billboard_fix3.csv")

#Publicar cantidad de datos por variable
print("Cantidad de datos: \n")
print(data.count())

#Publicar tipos de dato por variable
print("\nVariables: \n")
print(data.dtypes)

#Describir información acerca de la duración
print("\nDuración (segundos): ")

clean = []
#Limpiar datos para eliminar errores
for x in data.durationSeg:
    #Considerar solamente tiempos de menos de 500 segs y más de 0
    if (x<500 and x>0):
        clean.append(x)

print("Media: ",st.mean(clean))
print("Moda: ",st.mode(clean))
print("Desviación: ",st.stdev(clean))

#Graficar datos de duración
plt.title("Duración")
plt.plot(clean,"o")
plt.show()

#Mostrar rango de duración
print("Rango: ")
print(min(clean),"-",max(clean))


#Para cada una de las variables seleccionadas para describir
variables = ["artist","mood","tempo","genre"]

for variable in variables:
    
    
    print("\n"+variable+": ")
    repeticiones = {}
    
    #Contar cuántas veces se repite cada elemento
    for var in data[variable]:
        repeticiones.setdefault(var,0)
        repeticiones[var]+=1
    
    #Graficar repeticiones
    plt.title(variable)
    plt.bar(repeticiones.keys(),repeticiones.values())
    plt.show()
    
    
    
    #Acomodar diccionario de mayor a menor
    sorted_r = dict(sorted(repeticiones.items(), key=lambda item: item[1],reverse = True))
    x = 1
    
    #Publicar rango de repeticiones
    print("Rango de repeticiones: ",sorted_r[list(sorted_r)[-1]],"-",sorted_r[list(sorted_r)[0]],"\n")
    
    
    print("Primeros lugares: ")
    
    #Publicar únicamente los cinco primeros elementos, su posición, su nombre y su valor
    for key in sorted_r:
        print(x,".- ",key,": ",sorted_r[key])
        x+=1
        if(x>5):
            break
```
